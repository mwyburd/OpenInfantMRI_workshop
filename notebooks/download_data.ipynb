# Download open infant MRI data on OpenNeuro


There are many resources where open datasets are been published. 
OpenNEURO is one of the largest and ensures datasets follow the BIDS convention for easy use: https://bids.neuroimaging.io/index.html


## Step 1: Make an OpenNEURO account https://openneuro.org/

## Step 2: Search for Infant MRI datasets. 
Look for Infant FreeSurfer dataset (https://openneuro.org/datasets/ds004776/versions/1.0.0)

## Step 3: Explore the BIDS file structure

## Step 4: Download the data

There are a few ways to download the datasets. You can do it manually (by clicking the icons next to each file), through github or download the whole directory. We will try the 3rd options.

Click on Download, and following the *Download from S3* instructions. You will need aws-cli packaged. Install this in your environment

```python
import subprocess
subprocess.run("conda install conda-forge::awscli", shell=True, executable="/bin/bash")
```

```python
# copy the Download from S3 code. It will look something like this
# change the outpath to save into the open_data directory
outpath = "../open_data/OpenNeuro/" 
cmd =f"aws s3 sync --no-sign-request s3://openneuro.org/ds004776 {outpath}" 
subprocess.run(cmd, shell=True, executable="/bin/bash")
```
