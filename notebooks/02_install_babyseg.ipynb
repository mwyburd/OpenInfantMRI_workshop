# Installing babyseg CPU locally


BabySeg is a brain segmentation tool for infants and young children, designed to delineate anatomical structures in MRI without preprocessing. The tool can integrate information from multiple NIfTI image volumes of variable contrast, shape, and resolution in any order, provided that (1) their header geometries are correct, and (2) they are properly aligned in world space.

Read documentation here: https://github.com/mu40/babyseg/

## Step 1: Setting up installation


Lets start by starting up our BabySeg home:

````python
import subprocess
import sys
from pathlib import Path
import os

babyseg_home = (Path("..") / "babyseg").resolve()
````

## Step 2a: Installing on Linux or Mac

To install on mac or linux we are going to curl the program directly. 

You can use the bash commands:

```sh
curl -Lo babyseg get.babyseg.io
chmod +x babyseg
./babyseg -h
```

or the script below.

This should print out the help for babyseg installation. Read the installation

````python
env = os.environ.copy() # get the environment settings
env["BABYSEG_HOME"] = str(babyseg_home)
env["BABYSEG_SIF"] = str(babyseg_home)

subprocess.run(["curl", "-Lo", "babyseg", "get.babyseg.io"],env=env)
subprocess.run(["chmod", "+x", "babyseg"],env=env)
subprocess.run(["./babyseg", "-h"],env=env)
````

## Step 2b: Installing on Windows

For windows install, we will use a modified wrapper scripts

In the babyseg folder, you will see two wrapper scripts: wrapper.py and wrapper_windows.py. 
The wrapper script is provided by the babyseg developers: https://github.com/mu40/babyseg/blob/main/docker/wrapper.py, and we edited this for easy installation for windows.

**Requirements**
BabySeg provide a wrapper script that facilitates setup and use of BabySeg containers. 
It requires Python and supports any of the container platforms: Docker, Podman, Apptainer, or Singularity.

Running the code below will make our container file, e.g. babyseg_0.0.sif


````python
import os
import subprocess
import sys
from pathlib import Path

wrapper = babyseg_home / "wrapper_windows.py"

if not wrapper.exists():
    raise FileNotFoundError(f"wrapper.py not found at: {wrapper}")

env = os.environ.copy()
env["BABYSEG_HOME"] = str(babyseg_home)
env["BABYSEG_SIF"] = str(babyseg_home)

# Cross-platform: call the wrapper directly with the active Python interpreter.
result = subprocess.run(
    [sys.executable, str(wrapper), "-h"],
    env=env,
    text=True,
    capture_output=True,
    check=False,
 )

if result.stdout:
    print(result.stdout)
if result.returncode != 0:
    if result.stderr:
        print(result.stderr)
    raise RuntimeError(f"babyseg help command failed with exit code {result.returncode}")

# Make the script excutable
subprocess.run(["chmod", "+x", wrapper],env=env)
subprocess.run([wrapper, "-h"],env=env)
````
